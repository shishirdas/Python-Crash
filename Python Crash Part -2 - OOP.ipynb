### Inheritance

```python
## Inherit (single inheritance)
## Parent Class / Base class / Super class
class Car:
  def __init__(self, windows, door, enginetype) -> None:
     self.windows = windows
     self.door = door
     self.enginetype = enginetype


  def drive(self):
    print(f"The person will drive the {self.enginetype} card")
```

```python
## Derived / Child / subclass
class Tesla(Car):
  def __init__(self, windows, door, enginetype, is_selfDriving):
    super().__init__(windows, door, enginetype)
    self.is_selfDriving = is_selfDriving

tesla1 = Tesla(4, 4, "AB", True)
print(tesla1.windows)

```

```python
## Multiple Inheritance
## When a class inherits from more than one base class

## Base class 1
class Animal:
  def __init__(self, name) -> None:
    self.name = name

  def eat(self):
    print(f"{self.name} is eating from Animal Base Class")

## Base class 2
class Pet:
  def __init__(self, owner) -> None:
     self.owner = owner


# Derive class
class Dog(Animal, Pet):
  def __init__(self, name, owner) -> None:
    Animal.__init__(self, name)
    Pet.__init__(self, owner)


dog1 = Dog("ABC", "DEF")
dir(dog1)
```

### Polymorphism

```python
## Base Class
class Animal:
    def speak(self):
        return "Sound of the animal"

## Derived Class 1
class Dog(Animal):
    def speak(self):
        return "Woof!"

## Derived class 2
class Cat(Animal):
    def speak(self):
        return "Meow!"

cat1 = Cat()
print(cat1.speak())
dir(cat1)
```

### Encapsulation

```python
### Encapsulation  with Getter and Setter MEthods
### Public,protected,private variables or access modifiers

class Person:
    def __init__(self,name,age, balance):
        self.name=name    ## public variables
        self.age=age      ## public variables
        self.balance = balance


def get_name(person):
    return person.name

person1 = Person("Rakib", 20, 10000)
print(person1.balance)

person1.name
```

```python
# Getter and Setter MEthods
class Person:
    def __init__(self,name,age, balance):
        self.name=name    ## public variables
        self._age=age      ## protected variables -> Child class access
        self.__balance = balance ## private -> only nijer class

    def info(self):
      print("Balance ", self.__balance)

    def get_balance(self):
        return self.__balance

    def set_balance(self, money):
        self.__balance += money

# bayrer method
def get_name(person):
    return person.name

person1 = Person("Rakib", 20, 10000)
person1.set_balance(500)
print(person1.name)
print(person1.get_balance())
```

### Abstraction 

```python
from abc import ABC, abstractmethod

## Abstract base cclass
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")

    @abstractmethod
    def run(self):
        pass



class Car(Vehicle):
  def info(self):
    print("This is car")

  def start(self):
    print("Starting car engine")

  def ac(self):
    print("AC is started")

  def run(self):
    self.info()
    self.start()
    self.ac()


car1 = Car()
car1.run() ## Run swich
```
